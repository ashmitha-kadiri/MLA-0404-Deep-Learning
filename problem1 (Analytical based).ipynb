```python
x = 4
y = 15
W = 2
b = 1
learning_rate = 0.1

y_pred = W * x + b

loss = 0.5 * (y - y_pred) ** 2

dW = (y_pred - y) * x
db = y_pred - y

W_new = W - learning_rate * dW

b_new = b - learning_rate * db

print("Predicted output:", y_pred)
print("Loss:", loss)
print("Gradient of weight:", dW)
print("Gradient of bias:", db)
print("Updated weight:", W_new)
print("Updated bias:", b_new)
```
